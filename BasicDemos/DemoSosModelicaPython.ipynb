# Demo combined Modelica and Python notebooks

```sos
model testmodel3
 Real x;
equation
  der(x)=1-x;
end testmodel3;
```

```sos
simulate(testmodel3, stopTime=10)
```

```sos
plot(x,der(x))
```

```sos
import DyMat
res = DyMat.DyMatFile('testmodel3_res.mat')
```

```sos
def plot(x,y,z=None):
    %matplotlib inline
    import matplotlib.pyplot as plt
    fig = plt.figure()
    plt.plot(x,y,'r')
    if z is not None:
        plt.plot(x,z,'b')
    plt.show()
```

```sos
plot(res.abscissa('x',valuesOnly=True),res['x'],res['der(x)'])
```

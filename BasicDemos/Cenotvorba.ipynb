Cenotvorba pro projekt eGolem.

2 produkty

1. egolem.online - individuální cena 10 - 100 € ročně 250 - 2500 Kč
2. egolem - pro VŠ, simulační centra - 1000 - 20000 € ročně - 25 000 - 500 000 Kč.



```python
def plotPriceEstimate(x, fixedCost = 140000, variableCostPer1 = 100, pricePerPiece = 250):
        %matplotlib inline
        import matplotlib.pyplot as plt
        fig = plt.figure();        
        costs = [fixedCost + variableCostPer1 * i for i in x];
        gains = [pricePerPiece * i for i in x];
        gains2 = [pricePerPiece * 0.75 * i for i in x];
        gains3 = [pricePerPiece * 1.5 * i for i in x];
        x1 = fixedCost / (pricePerPiece - variableCostPer1)
        x2 = fixedCost / (pricePerPiece * 0.75 - variableCostPer1)
        x3 = fixedCost / (pricePerPiece * 1.5 - variableCostPer1)
        plt.plot(x,costs,'r')
        plt.plot(x,gains,'b')
        plt.plot(x,gains2,'b--')
        plt.plot(x,gains3,'b--')
        plt.axvline(x = x3, color = 'g', linestyle='dotted', label=f"{x3:.1f} for the price {pricePerPiece*1.5}")
        plt.axvline(x = x1, color = 'g', linestyle='dotted', label=f"{x1:.1f} for the price {pricePerPiece}")
        plt.axvline(x = x2, color = 'g', linestyle='dotted', label=f"{x2:.1f} for the price {pricePerPiece*0.75}")                
        plt.legend()
        plt.show()
        print("price:",pricePerPiece, " price 1/2:",pricePerPiece*0.5," price 3/2: ",pricePerPiece*1.5)
        
    
```

# 1. Egolem.online - minimální tým
2 grafici, 1 sales, 1 vědec, 1 programátor a 1 tvůrce obsahu na 1/2 úvazek
Měsíční náklady na minimální tým jsou 140 000 Kč.

```python
plotPriceEstimate(range(0,1200,10),fixedCost = 140000, pricePerPiece = 250)
```

Za cenu 250 Kč, nebo €10 se dosáhne bodu zlomu u 933 předplatitelů měsíčně. Tj. tým si na sebe vydělá.

Za ceny 375 Kč, nebo €15 je bod zlomu 509 předplatitelů.

# 1.2 Egolem.online - Plný tým
Pokud se dosáhne bodu zlomu v minimálním týmu, může se začít najímat další lidi do plného týmu: 2 grafici, 2 programátoři, 2 vědci/lektoři, 1 administrátor na plné úvazky (30 - 75 tis Kč. měsíčně) stojí asi 700 000 Kč.


```python
plotPriceEstimate(range(0,10000,100),fixedCost = 700000, pricePerPiece = 250)
```

Za cenu 250 Kč, nebo €10 je bod zlomu 4666 předplatitelů měsíčně.

Za cenu 375 Kč, nebo €15 je bod zlomu 2545 předplatitelů.

# 2. eGolem - minimální tým
2 grafici, 1 sales, 1 vědec, 1 programátor a 1 tvůrce obsahu na 1/2 úvazek Měsíční náklady na minimální tým jsou 140 000 Kč.

```python
plotPriceEstimate(range(0,10,1),fixedCost = 140000*12, pricePerPiece = 25000*12, variableCostPer1 = 1000*12)
```
